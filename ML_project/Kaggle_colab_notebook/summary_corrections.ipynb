```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/summarized_data.csv')
df.head()
```

```python
df['All_Reviews'] = df['All_Reviews'].str.replace('na', '', regex=False)
df['All_Reviews'] = df['All_Reviews'].str.replace('error', '', regex=False)
df['All_Reviews'] = df['All_Reviews'].str.replace('NaN', '', regex=False)
df['All_Reviews'] = df['All_Reviews'].str.strip()
df['All_Reviews'] = df['All_Reviews'].replace([''], 'No reviews available for this mobile')
df['All_Reviews'] = df['All_Reviews'].fillna('No reviews available for this mobile')
df['All_Reviews'] = df['All_Reviews'].replace('', 'No reviews available for this mobile')
mask = df['All_Reviews'].eq('')
df.loc[mask, 'All_Reviews'] = 'No reviews available for this mobile'
df.head()
```

```python
df = df[['Product Name','All_Reviews','summary']]
df.head()
```

```python
df.loc[df['All_Reviews'] == 'No reviews available for this mobile', 'summary'] = 'No reviews available for this mobile'
```

```python
df1 = df.iloc[2000:2010, :]
df1
```

```python
!pip install transformers
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load a grammar correction model
model_name = "t5-small"  # or a fine-tuned model like "t5-base-finetuned-grammar-correction"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def correct_grammar(text):
    if not isinstance(text, str) or text.strip() == "":
        return text
    input_text = f"correct: {text}"
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(inputs, max_length=512, num_beams=5, early_stopping=True)
    corrected_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return corrected_text

# Apply grammar correction to the summary column
df1['summary'] = df1['summary'].apply(correct_grammar)
```

```python
from textblob import TextBlob

def has_error(text):
    blob = TextBlob(text)
    return blob.correct() != text

# Filter rows with errors
df1['has_error'] = df1['summary'].apply(has_error)
error_rows = df1[df1['has_error']]
```
