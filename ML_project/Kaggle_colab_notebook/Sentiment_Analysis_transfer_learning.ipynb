```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/combined_data.csv")
pd.set_option('display.max_colwidth', None)
df.head(1)
```

```python
review_columns = ['Camera_Comments', 'Battery_Comments', 'Display_Comments', 'Design_Comments', 'Performance_Comments']
df['All_Reviews'] = df[review_columns].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
df = df[["Product Name", "All_Reviews"]]
df.tail(1)
```

```python
import re
import string
df["All_Reviews"] = df["All_Reviews"].str.lower()
def remove_html_tag(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub('', text)
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', text)
def remove_newline(text):
    return text.replace('\n', ' ')
def remove_white_space(text):
    return re.sub(r'\s+', ' ', text)
def remove_punc(text):
    exclude = string.punctuation
    return text.translate(str.maketrans('', '', exclude))
df["All_Reviews"] = df["All_Reviews"].apply(remove_html_tag)
df["All_Reviews"] = df["All_Reviews"].apply(remove_newline)
df["All_Reviews"] = df["All_Reviews"].apply(remove_url)
df["All_Reviews"] = df["All_Reviews"].apply(remove_white_space)
df["All_Reviews"] = df["All_Reviews"].apply(remove_punc)
df.head(1)
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["  # Start of character set
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # Chinese characters
        u"\U00002702-\U000027B0"  # Miscellaneous symbols
        u"\U000024C2-\U0001F251"  # Enclosed characters
        u"\U00010000-\U0010FFFF"  # Supplementary characters
        "]+", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)
df["All_Reviews"] = df["All_Reviews"].apply(remove_emoji)
first_row = df.iloc[0]
print(first_row)
```

```python
import pandas as pd
import torch
from transformers import BartForConditionalGeneration, BartTokenizer
from tqdm import tqdm

device = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name).to(device)
tokenizer = BartTokenizer.from_pretrained(model_name)

if "All_Reviews" not in df.columns:
    raise ValueError("The DataFrame must contain an 'All_Reviews' column.")

if "Summary" not in df.columns:
    df["Summary"] = None

output_file = "/content/drive/MyDrive/review_summaries.csv"

start_row = 0
try:

    saved_df = pd.read_csv(output_file)
    start_row = len(saved_df)
    df.loc[:start_row - 1, "Summary"] = saved_df["Summary"]
    print(f"Resuming from row {start_row}")
except FileNotFoundError:
    print("No saved progress found. Starting fresh.")

for i in tqdm(range(start_row, len(df))):
    review = df.loc[i, "All_Reviews"]

    if pd.isna(review) or len(review.strip()) == 0:
        summary = "No review available"
    else:
        try:

            inputs = tokenizer.encode("summarize: " + review, return_tensors="pt", truncation=True, max_length=1024).to(device)
            summary_ids = model.generate(inputs, max_length=130, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
            summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        except Exception as e:
            summary = f"Error generating summary: {e}"

    df.loc[i, "Summary"] = summary

    df.iloc[:i + 1].to_csv(output_file, index=False)

print("Summary generation completed.")
```
