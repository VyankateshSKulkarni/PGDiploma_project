```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/sentiment_data_cleaned.csv')
df.head()
```

```python
df.rename(columns={'Product Name':'Mobile_Name'},inplace=True)
```

```python
df.loc[df['Mobile_Name'].str.contains('Apple iPhone 16', case=False), 'Sentiment_Label'] = 'POSITIVE'
```

```python
POSITIVE =[]
NEGATIVE =[]
for i in range(len(df)):
  if df['Sentiment_Label'][i] == 'POSITIVE':
    POSITIVE.append(float(df['Sentiment_Score'][i]))
    NEGATIVE.append(1 - float(df['Sentiment_Score'][i]))
  else:
    POSITIVE.append(1- float(df['Sentiment_Score'][i]))
    NEGATIVE.append(float(df['Sentiment_Score'][i]))

df['Positive_score'] = POSITIVE
df['Negative_score'] = NEGATIVE
```

```python
df.drop(['All_Reviews','Sentiment_Label', 'Sentiment_Score'], axis=1, inplace=True)
df.head()
```
