```python
import pandas as pd
import re

# Function to extract ratings and reviews
def extract_ratings_reviews(row):
    # Extract ratings
    ratings_match = re.search(r'([\d,]+) Ratings', row)
    ratings = int(ratings_match.group(1).replace(',', '')) if ratings_match else 0
    
    # Extract reviews
    reviews_match = re.search(r'([\d,]+) Reviews', row)
    reviews = int(reviews_match.group(1).replace(',', '')) if reviews_match else 0
    
    return pd.Series({'Rating_count': ratings, 'Review_count': reviews})

# # Example DataFrame
# data = {'Details': ['1,39,977 Ratings & 4,510 Reviews', '5,000 Ratings & 350 Reviews']}
# df = pd.DataFrame(data)

# Apply the function to create new columns
df[['Rating_count', 'Review_count']] = df['Rating Count'].apply(extract_ratings_reviews)

df.head(2)
```

```python
col_of_overall_ratings = ["Camera_Overall_Rating",'Battery_Overall_Rating','Display_Overall_Rating','Design_Overall_Rating',
                         'Performance_Overall_Rating']
def overall_ratings_num(col):
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Ensure numeric
    df[col] = df[col].fillna(0.5)  # Replace NaN with 0 (or another strategy)
    df[col] = df[col] / 5
for col in col_of_overall_ratings:
    overall_ratings_num(col)
df.head(2)
```

```python
def score_num(col):
    df[col] = df[col].str.replace('%','', regex=False) 
    df[col] = pd.to_numeric(df[col], errors='coerce') 
    df[col] = df[col].fillna(50) 
    df[col] = df[col] / 100 
list_col = [
    "Camera_Positive_Score",
    "Camera_Negative_Score",
    "Battery_Positive_Score",
    "Battery_Negative_Score",
    "Display_Positive_Score",
    "Display_Negative_Score",
    "Design_Positive_Score",
    "Design_Negative_Score",
    "Performance_Positive_Score",
    "Performance_Negative_Score"
]
def calculate_missing_score(row):
    # Count 0.5 values only in these specific columns
    count_0_5 = sum(row[col] == 0.5 for col in list_col)
    
    if count_0_5 == 0:
        return 1.0
    elif count_0_5 == 1:
        return 0.7
    elif count_0_5 == 2:
        return 0.6
    elif count_0_5 == 3:
        return 0.5
    elif count_0_5 == 4:
        return 0.4
    else:
        return 0.3

# Apply the score_num function to all columns
for col in list_col:
    score_num(col)

# Calculate the Any_Score_Missing values based on only the 8 specific columns
df['Any_Score_Missing'] = df.apply(calculate_missing_score, axis=1)
```

```python
def star_na_fill(col):
    df[col] = df[col].fillna(0)
star_col = ["5_star","4_star","3_star","2_star","1_star"]
for col  in star_col:
    star_na_fill(col)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
data = pd.read_csv("numeric.csv")  # Replace with your file
features = [
    'Rating', 'Rating_counts', 'Review_counts', 'Price',
    '5_star', '4_star', '3_star', '2_star', '1_star',
    'Camera_Overall_Rating', 'Camera_Positive_Score', 'Camera_Negative_Score',
    'Battery_Overall_Rating', 'Battery_Positive_Score', 'Battery_Negative_Score',
    'Display_Overall_Rating', 'Display_Positive_Score', 'Display_Negative_Score',
    'Design_Overall_Rating', 'Design_Positive_Score', 'Design_Negative_Score',
    'Performance_Overall_Rating', 'Performance_Positive_Score', 'Performance_Negative_Score', 'Any_Score_Missing'
]
similarity_matrix = cosine_similarity(data[features])
similarity_df = pd.DataFrame(similarity_matrix)
similarity_df.to_csv('similarity_matrix.csv', index=False)
similarity_df.head(2)
```

```python
def recommend_mobile(user_input, df, similarity_df):
    unique_names = df['Mobile_name'].unique()
    if user_input not in unique_names:
        print("Mobile name not found.")
        return []
    index = df[df['Mobile_name'] == user_input].index[0]
    similarity_scores = similarity_df.iloc[index]
    top_indices = similarity_scores.nlargest(11).index[1:]  
    recommended_mobiles = df.iloc[top_indices]['Mobile_name'].tolist()
    recommended_set = set(recommended_mobiles)
    recommended_set.discard(user_input) 
    return list(recommended_set)[:5]
    
user_input = 'Infinix Smart 7 HD (Green Apple, 64 GB)'
recommendations = recommend_mobile(user_input, data, similarity_df)
print(recommendations)
```

```python
i = 0
for mobile in data['Mobile_name'][600:]:
    rec = recommend_mobile(mobile, data, similarity_df)
    print(f"{mobile} :: {rec[0]}")
    print(f"2) :: {rec[1]}")
    print(f"3) :: {rec[2]}")
    print(f"4) :: {rec[3]}")
    print(f"5) :: {rec[4]}")
    i=i+1
    if i> 10:
        break
```
