```python
import numpy as np
import pandas as pd 
df = pd.read_csv("apple_with_ratings_and_feature_hrefs.csv")
pd.set_option('display.max_colwidth', None)
df.head(1)
```

```python
df = df[['Product Name','Camera_Comments','Battery_Comments','Display_Comments','Design_Comments','Performance_Comments']]
df.head(1)
```

```python
review_columns = ['Camera_Comments', 'Battery_Comments', 'Display_Comments', 'Design_Comments', 'Performance_Comments']
df['All_Reviews'] = df[review_columns].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
df = df[["Product Name", "All_Reviews"]]
df.tail(1)
```

```python
df.tail()
```

```python
import re
import string
df["All_Reviews"] = df["All_Reviews"].str.lower()
def remove_html_tag(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub('', text)
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', text)
def remove_newline(text):
    return text.replace('\n', ' ')
def remove_white_space(text):
    return re.sub(r'\s+', ' ', text)
def remove_punc(text):
    exclude = string.punctuation
    return text.translate(str.maketrans('', '', exclude))
df["All_Reviews"] = df["All_Reviews"].apply(remove_html_tag) 
df["All_Reviews"] = df["All_Reviews"].apply(remove_newline)
df["All_Reviews"] = df["All_Reviews"].apply(remove_url) 
df["All_Reviews"] = df["All_Reviews"].apply(remove_white_space)
df["All_Reviews"] = df["All_Reviews"].apply(remove_punc)
df.head(1)
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["  # Start of character set
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # Chinese characters
        u"\U00002702-\U000027B0"  # Miscellaneous symbols
        u"\U000024C2-\U0001F251"  # Enclosed characters
        u"\U00010000-\U0010FFFF"  # Supplementary characters
        "]+", flags=re.UNICODE)
    
    return emoji_pattern.sub(r'', text)
df["All_Reviews"] = df["All_Reviews"].apply(remove_emoji)
first_row = df.iloc[0]
print(first_row)
```

```python
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Sample input text (after initial cleaning)
cleaned_text = """the camera quality is exceptional capturing detailed and vivid photos even in low light..."""

# Step 1: Tokenization
nltk.download('punkt')
tokens = word_tokenize(cleaned_text)

# Step 2: Stop Words Removal
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Step 3: Stemming (or Lemmatization)
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

# Step 4: Removing Numbers (if applicable)
cleaned_text = re.sub(r'\d+', '', cleaned_text)

# Step 5: Removing Extra Whitespace
cleaned_text = ' '.join(cleaned_text.split())

# Resulting tokens after preprocessing steps
print(stemmed_tokens)
```
