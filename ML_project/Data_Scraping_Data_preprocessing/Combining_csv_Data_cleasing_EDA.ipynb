```python
import pandas as pd
import re

# Function to extract ratings and reviews
def extract_ratings_reviews(row):
    # Extract ratings
    ratings_match = re.search(r'([\d,]+) Ratings', row)
    ratings = int(ratings_match.group(1).replace(',', '')) if ratings_match else 0
    
    # Extract reviews
    reviews_match = re.search(r'([\d,]+) Reviews', row)
    reviews = int(reviews_match.group(1).replace(',', '')) if reviews_match else 0
    
    return pd.Series({'Rating_count': ratings, 'Review_count': reviews})

# # Example DataFrame
# data = {'Details': ['1,39,977 Ratings & 4,510 Reviews', '5,000 Ratings & 350 Reviews']}
# df = pd.DataFrame(data)

# Apply the function to create new columns
df[['Rating_count', 'Review_count']] = df['Rating Count'].apply(extract_ratings_reviews)

df.head(2)
```

```python
# def score_num(col):
#     df[col] = df[col].str.replace('%','', regex=False) 
#     df[col] = pd.to_numeric(df[col], errors='coerce') 
#     df[col] = df[col].fillna(50) 
#     df[col] = df[col] / 100 
# list_col = [
#     "Camera_Positive_Score",
#     "Camera_Negative_Score",
#     "Battery_Positive_Score",
#     "Battery_Negative_Score",
#     "Display_Positive_Score",
#     "Display_Negative_Score",
#     "Design_Positive_Score",
#     "Design_Negative_Score",
#     "Performance_Positive_Score",
#     "Performance_Negative_Score"
# ]
# for col in list_col:
#     score_num(col)
# df['Any_Score_Missing'] = df[list_col].apply(lambda row: 1 if (row == 0.5).any() else 0, axis=1)
# df.head(2)
def score_num(col):
    df[col] = df[col].str.replace('%','', regex=False) 
    df[col] = pd.to_numeric(df[col], errors='coerce') 
    df[col] = df[col].fillna(50) 
    df[col] = df[col] / 100 
list_col = [
    "Camera_Positive_Score",
    "Camera_Negative_Score",
    "Battery_Positive_Score",
    "Battery_Negative_Score",
    "Display_Positive_Score",
    "Display_Negative_Score",
    "Design_Positive_Score",
    "Design_Negative_Score",
    "Performance_Positive_Score",
    "Performance_Negative_Score"
]
def calculate_missing_score(row):
    # Count 0.5 values only in these specific columns
    count_0_5 = sum(row[col] == 0.5 for col in list_col)
    
    if count_0_5 == 0:
        return 1.0
    elif count_0_5 == 1:
        return 0.7
    elif count_0_5 == 2:
        return 0.6
    elif count_0_5 == 3:
        return 0.5
    elif count_0_5 == 4:
        return 0.4
    else:
        return 0.3

# Apply the score_num function to all columns
for col in list_col:
    score_num(col)

# Calculate the Any_Score_Missing values based on only the 8 specific columns
df['Any_Score_Missing'] = df.apply(calculate_missing_score, axis=1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
# Compute the cosine similarity matrix
temp = numeric_df.iloc[:,1:]
similarity_matrix = cosine_similarity(temp)

# Convert the similarity matrix into a DataFrame for easier interpretation
similarity_df = pd.DataFrame(similarity_matrix, index=df["Product Name"], columns=df["Product Name"])
```

```python
def recommend_products(product_name, similarity_df, top_n=5):
    # Get the similarity scores for the given product
    similar_scores = similarity_df[product_name].sort_values(ascending=False)
    
    # Exclude the product itself (i.e., the first value in the similarity score)
    similar_scores = similar_scores.drop(product_name)
    
    # Get the top N similar products
    top_recommendations = similar_scores.head(top_n)
    
    return top_recommendations
```

```python
recommend_products("Nothing Phone (2a) Plus (Black, 256 GB)", similarity_df, top_n=5)
```
