```python
import re
import string
df["All_Reviews"] = df["All_Reviews"].str.lower()
def remove_html_tag(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub('', text)
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub('', text)
def remove_newline(text):
    return text.replace('\n', ' ')
def remove_white_space(text):
    return re.sub(r'\s+', ' ', text)
def remove_punc(text):
    exclude = string.punctuation
    return text.translate(str.maketrans('', '', exclude))
def remove_emoji(text):
    emoji_pattern = re.compile(
        "["  
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # Chinese characters
        u"\U00002702-\U000027B0"  # Miscellaneous symbols
        u"\U000024C2-\U0001F251"  # Enclosed characters
        u"\U00010000-\U0010FFFF"  # Supplementary characters
        "]+", flags=re.UNICODE)
    
    return emoji_pattern.sub(r'', text)    


df["All_Reviews"] = df["All_Reviews"].apply(remove_html_tag) 
df["All_Reviews"] = df["All_Reviews"].apply(remove_newline)
df["All_Reviews"] = df["All_Reviews"].apply(remove_url) 
df["All_Reviews"] = df["All_Reviews"].apply(remove_white_space)
df["All_Reviews"] = df["All_Reviews"].apply(remove_punc)
df["All_Reviews"] = df["All_Reviews"].apply(remove_emoji)
```

```python
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob


# Function to correct spelling using TextBlob
# def correct_spelling_with_textblob(text):
#     blob = TextBlob(text)
#     return blob.correct()
    
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    return ' '.join([word for word in text.split() if word not in stop_words])

# # 4. Tokenization (split the text into individual words)
# def tokenize_text(text):
#     tokens = text.split()
#     return tokens
    

# 6. Lemmatization
def lemmatization(text):
    lemmatizer = WordNetLemmatizer()
    return ' '.join([lemmatizer.lemmatize(word,pos='v') for word in text.split()])


df['All_Reviews'] = df['All_Reviews'].apply(remove_stopwords)
# df['All_Reviews'] = df['All_Reviews'].apply(tokenize_text)
df['All_Reviews'] = df['All_Reviews'].apply(lemmatization)

```

```python
from transformers import pipeline

# Initialize the sentiment-analysis pipeline with truncation and padding
sentiment_analyzer = pipeline('sentiment-analysis', truncation=True, padding=True, max_length=512)

# Apply sentiment analysis to the reviews and include both label and score
df['Sentiment'] = df['All_Reviews'].apply(lambda x: sentiment_analyzer(x)[0])

# Split the resulting dictionary into separate 'Sentiment_Label' and 'Sentiment_Score' columns
df['Sentiment_Label'] = df['Sentiment'].apply(lambda x: x['label'])
df['Sentiment_Score'] = df['Sentiment'].apply(lambda x: x['score'])

# Drop the original 'Sentiment' column (optional)
df.drop(columns=['Sentiment'], inplace=True)

# Display the updated DataFrame
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000)
x = tfidf.fit_transform(df['All_Reviews']).toarray()
y = df['Sentiment_Label']
```

```python
# split the data into train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=123456)
```

```python
from sklearn.linear_model import LogisticRegression
def create_LR_model():
    model = LogisticRegression(max_iter=1000)
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.naive_bayes import GaussianNB
def create_NB_model():
    model = GaussianNB()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier
def create_KNN_model():
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.svm import SVC
def create_SVM_model():
    model = SVC()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.tree import DecisionTreeClassifier
def create_DT_model():
    model = DecisionTreeClassifier(max_depth=500)
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.ensemble import RandomForestClassifier
def create_RF_model():
    model = RandomForestClassifier(n_estimators=100)
    model.fit(x_train,y_train)
    return model
```

```python
from catboost import CatBoostClassifier
def create_CatBoost_model():
    model = CatBoostClassifier(logging_level='Silent')
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.ensemble import AdaBoostClassifier
def create_Adaboost_model():
    model = AdaBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.ensemble import GradientBoostingClassifier
def create_GB_model():
    model = GradientBoostingClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
from xgboost import XGBClassifier
def create_XGB_model():
    model = XGBClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
models = [
    ("Logistic Regression",create_LR_model(),(0,0)),
    ("Naive Bayes",create_NB_model(),(0,1)),
    ("KNN",create_KNN_model(),(1,0)),
    ("SVM",create_SVM_model(),(1,1)),
    ("Decision Tree",create_DT_model(),(2,0)),
    ("Random Forest",create_RF_model(),(2,1)),
    ("CatBoost",create_CatBoost_model(),(3,0)),
    ("Adaboost",create_Adaboost_model(),(3,1)),
    ("Gradient Boost",create_GB_model(),(4,0)),
    ("XG Boost",create_XGB_model(),(4,1))
]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
def evaluate_model(model):
    y_pred = model.predict(x_test)
    y_true = y_test

    cm = confusion_matrix(y_true,y_pred)
    accuracy = accuracy_score(y_true,y_pred)
    precision = precision_score(y_true,y_pred,average='micro')
    recall = recall_score(y_true,y_pred,average='micro')
    f1 = f1_score(y_true,y_pred,average='micro')

    return cm,accuracy,precision,recall,f1
```

```python
performance_chart_data = []
for index in models:
    # y_pred = model.predict(x_test)
    y_true = y_test
    cm,accuracy,precision,recall,f1 = evaluate_model(index[1])
    performance_chart_data.append([index[0], cm,accuracy,precision,recall,f1])
    performance_chart = pd.DataFrame(performance_chart_data,
                                    columns=["Model name","Confusion Matrix","Accuracy","Precision","Recall","F1 Score"])

performance_chart
```
